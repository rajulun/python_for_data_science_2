***Вопросы по теоретической части урока 1:***

***Задача 1. Привести по 2 примера, когда лучше максимизировать Precision, а когда Recall?***


Все зависит от поставленных задач.

Точность (precision) — показывает, какой процент из предсказанных позитивных объектов, оказались действительно позитивными.

1. В банке, при оценке кредитного риска. Если перед нами стоит задача минимизировать риск лучше максимизировать Precision 

2. Например, при отборе нянек по фотографии. Отбираем только тех, кого наша модель пометила только истино-положительными


Полнота (recall) это доля правильно распознаных объектов от всех целевых объектов.
Приоритет на Полноту необходим в том случае, когда пропуск целевого объекта критичен, при этом отнесение неверного объекта к целевому не является критической ошибкой, т.к. после можно провести дополнительную проверку.

1. Максимизировать recall можно, например, в медицине. Здесь важно зафиксировать не только болезнь, но и подозрение на болезнь. В случае подозрения направить на дополнительное обследование. 

2. На производстве. В аналогии с первым примером отбраковку можно направить на ручную проверку

***Задача 2. Почему мы используем F-меру, почему, например, нельзя просто взять среднее от Precision и Recall?***

В метрике F-score в результе мы получаем гармоническое среднее между precision и recall. В зависимости от поставленных задач, мы можем балансировать вкладом, вносимым вышеуказанными параметрами. Таким образом мы акцентируем внимание на том параметре, который более предпочтителен для нашей задачи.

Среднее от Precision и Recall подходит далеко не под все задачи
